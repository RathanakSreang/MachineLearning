```python
from keras.datasets import cifar10
import matplotlib.pyplot as plt
import scipy.misc as misc
%matplotlib inline  
#load data
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
# create a grid of 3x3 images
for i in range(0,9):
    plt.subplot(330 + 1 + i)
    plt.imshow(misc.toimage(X_train[i]))
#show the plot
plt.show()
```
